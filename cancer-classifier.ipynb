```python
# using cross validation to reduce over fitting
# Using semi supervised learning method self_training for unlabeled data
# Using XGBoost model for classification
# Obesreving graphs, best value to fill is mean value
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, f1_score
from xgboost import XGBClassifier
from sklearn.semi_supervised import SelfTrainingClassifier

X_labeled = train.loc[train['Class'].notna(), train.columns.difference(['Id', 'Class'])]
y_labeled = train.loc[train['Class'].notna(), 'Class']

# implementing the cross validation loop
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []
f1s = []

for fold, (train_idx, val_idx) in enumerate(kf.split(X_labeled, y_labeled), 1):
    print(f"\nðŸš€ Fold {fold}")
    
    # Split
    X_train_raw = X_labeled.iloc[train_idx]
    X_val_raw = X_labeled.iloc[val_idx]
    y_train = y_labeled.iloc[train_idx]
    y_val = y_labeled.iloc[val_idx]
    
    # Impute with train mean
    mean_vals = X_train_raw.mean()
    X_train_filled = X_train_raw.fillna(mean_vals)
    X_val_filled = X_val_raw.fillna(mean_vals)
    # Prepare unlabeled data
    X_unlabeled = train.loc[train['Class'].isna(), train.columns.difference(['Id', 'Class'])]
    X_unlabeled_filled = X_unlabeled.fillna(mean_vals)
    y_unlabeled = np.full(len(X_unlabeled), -1)

    # Combine
    X_combined = pd.concat([X_train_filled, X_unlabeled_filled], ignore_index=True)
    y_combined = pd.concat([y_train.reset_index(drop=True), pd.Series(y_unlabeled)], ignore_index=True)

    # Model
    base_clf = XGBClassifier(
        tree_method='gpu_hist',    
        objective='multi:softprob',
        num_class=len(np.unique(y_labeled)),
        use_label_encoder=False,
        eval_metric='mlogloss',
        random_state=42,
        n_jobs=-1,
        n_estimators=50,              
        max_depth=3
        )

    self_training_clf = SelfTrainingClassifier(
        base_clf,
        threshold=0.9,
        criterion='k_best',
        k_best=10,
        max_iter=1000,
        verbose=False
        )

    # Fit model
    self_training_clf.fit(X_combined, y_combined)
    # Evaluate
    y_pred = self_training_clf.predict(X_val_filled)
    acc = accuracy_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred, average='macro')

    print(f"ðŸ”¹ Fold {fold} Accuracy: {acc:.4f}")
    print(f"ðŸ”¹ Fold {fold} Macro F1 Score: {f1:.4f}")

    accuracies.append(acc)
    f1s.append(f1)
```

```python
# final training
from xgboost import XGBClassifier
from sklearn.semi_supervised import SelfTrainingClassifier
import numpy as np
import pandas as pd

# Preparing full labeled and unlabeled data ---
X_labeled_final = train.loc[train['Class'].notna(), train.columns.difference(['Id', 'Class'])]
y_labeled_final = train.loc[train['Class'].notna(), 'Class']

X_unlabeled_final = train.loc[train['Class'].isna(), train.columns.difference(['Id', 'Class'])]
y_unlabeled_final = np.full(len(X_unlabeled_final), -1)

# Filling NaNs using only labeled training stats ---
mean_final = X_labeled_final.mean()
X_labeled_filled = X_labeled_final.fillna(mean_final)
X_unlabeled_filled = X_unlabeled_final.fillna(mean_final)

# Combining for semi-supervised training ---
X_combined_final = pd.concat([X_labeled_filled, X_unlabeled_filled], ignore_index=True)
y_combined_final = pd.concat(
    [y_labeled_final.reset_index(drop=True), pd.Series(y_unlabeled_final)],
    ignore_index=True
)

# Defining and training final model (GPU-accelerated)
base_clf_final = XGBClassifier(
    tree_method='gpu_hist',       
    objective='multi:softprob',
    num_class=len(np.unique(y_labeled_final)),
    use_label_encoder=False,
    eval_metric='mlogloss',
    random_state=42,
    n_jobs=-1,
    n_estimators=50,
    max_depth=3
)

self_training_clf_final = SelfTrainingClassifier(
    base_clf_final,
    threshold=0.9,
    criterion='k_best',
    k_best=10,
    max_iter=1000,
    verbose=True
)

# fitting
self_training_clf_final.fit(X_combined_final, y_combined_final)
```
